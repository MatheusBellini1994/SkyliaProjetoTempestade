```python
import pandas as pd
import ast
```

```python
dfBruto = pd.read_csv('dadosCGEbruto.csv')
dfBruto
```

```python
dfBruto.info()
```

```python
dfTratado = pd.DataFrame()
dfTratado['Data'] = []
dfTratado['Zona'] = []
dfTratado['Bairro'] = []
dfTratado['Ponto'] = []
dfTratado['Status'] = []
dfTratado['Sentido'] = []
dfTratado['Referencia'] = []
dfTratado['HorarioInicio'] = []
dfTratado['HorarioFinal'] = []
dfTratado['Logradouro'] = []
dfTratado
```

```python
for i in range(len(dfBruto)):
  astdata = ast.literal_eval(dfBruto['data'][i])
  for itemastdata in astdata:
    for alerta in itemastdata.get('info').get('alertas'):
      linha = []
      linha.append(dfBruto['date'][i])
      linha.append(itemastdata.get('zona'))
      linha.append(itemastdata.get('info').get('bairro'))
      linha.append(itemastdata.get('info').get('ponto'))
      linha.append(alerta.get('status'))
      linha.append(alerta.get('sentido'))
      linha.append(alerta.get('referencia'))
      linha.append(alerta.get('horario').get('inicio'))
      linha.append(alerta.get('horario').get('final'))
      linha.append(alerta.get('logradouro'))
      dfTratado.loc[len(dfTratado)]=linha
```
