```python
dfmelted['Cód. Bairros'] = [x[:2] for x in dfmelted['Bairros']]
dfmelted
```

```python
dfmelted['Nome Bairros'] = [x[5:] for x in dfmelted['Bairros']]
dfmelted
```

```python
dfmelted = dfmelted[['Cód. Bairros','Nome Bairros','Dia','Pluviometria']]
dfmelted
```
